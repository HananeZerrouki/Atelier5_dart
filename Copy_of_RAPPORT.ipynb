```python
import matplotlib.pyplot as plt
# 1. Préparation des données (Simulation basée sur ton AQI de 27)
heures = ['08:00', '10:00', '12:00', '14:00', '16:00']
aqi_montreal = [22, 27, 35, 30, 45]
# 2. Création du graphique
plt.figure(figsize=(10, 5))
plt.plot(heures, aqi_montreal, marker='o', linestyle='-', color='skyblue', linewidth=3)
# 3. Ajout des détails (Titres et Légendes)
plt.title("Évolution de la Qualité de l'Air à Montréal (AQI)", fontsize=14)
plt.xlabel("Heure de la journée")
plt.ylabel("Indice de pollution (AQI)")
plt.grid(True, linestyle='--', alpha=0.6)
# 4. Ajout de la ligne de sécurité (AQI 50)
plt.axhline(y=50, color='red', linestyle='--', label='Seuil de qualité "Bonne" (50)')
plt.legend()
# 5. Affichage
plt.show()
```
